## An easy (but fun) derivation

Let us consider the arbitrary direction
$$
\hat{n} = \sin\theta\cos\phi \,\hat{x} + \sin\theta\sin\phi \,\hat{y} + \cos\theta \,\hat{z}.
$$
Its operator matrix $S_n$ is
\begin{align*}
    S_n &= \vec{S} \cdot \hat{n} \\
    \\
    &= \frac{\hbar}{2} \left[
    \sin\theta\cos\phi
    \begin{pmatrix}
    0 & 1 \\
    1 & 0
    \end{pmatrix}
    + \sin\theta\sin\phi
    \begin{pmatrix}
    0 & -i \\
    i & 0
    \end{pmatrix}
    + \cos\theta
    \begin{pmatrix}
    1 & 0 \\
    0 & -1
    \end{pmatrix}
    \right] \\
    \\
    &= \frac{\hbar}{2}
    \begin{pmatrix}
    \cos\theta & \sin\theta\,\mathrm{e}^{-i\phi} \\
    \sin\theta\,\mathrm{e}^{i\phi} & -\cos\theta
    \end{pmatrix}.
\end{align*}
L'équation caractéristique est
\begin{equation*}
    \begin{vmatrix}
    \frac{\hbar}{2}\cos\theta-\lambda & \frac{\hbar}{2}\sin\theta\,\mathrm{e}^{-i\phi} \\
    \frac{\hbar}{2}\sin\theta\,\mathrm{e}^{i\phi} & -\frac{\hbar}{2}\cos\theta-\lambda
    \end{vmatrix}
    = 0.
\end{equation*}
Le calcul du déterminant donne
$$
\lambda = \pm \frac{\hbar}{2}.
$$
Les fonctions propres sont obtenues selon
\begin{equation*}
    \frac{\hbar}{2}
    \begin{pmatrix}
    \cos\theta & \sin\theta\,\mathrm{e}^{-i\phi} \\
    \sin\theta\,\mathrm{e}^{i\phi} & -\cos\theta
    \end{pmatrix}
    \begin{pmatrix}
    \alpha \\
    \beta
    \end{pmatrix}
    = \pm \frac{\hbar}{2}
    \begin{pmatrix}
    \alpha \\
    \beta
    \end{pmatrix},
\end{equation*}
et la multiplication matricielle nous donne la relation
$$
\frac{\hbar}{2} \left(\alpha \cos\theta + \beta \sin\theta\,\mathrm{e}^{-i\phi}\right) = \pm \frac{\hbar}{2} \alpha.
$$
Pour la fonction propre $\chi_+^{(n)}$ (dont la valeur propre associée est $+\frac{\hbar}{2}$, soit celle désirée), la relation devient
$$
\beta = \alpha \frac{1 - \cos\theta}{\sin\theta} \mathrm{e}^{i\phi} = \alpha \tan\left(\frac{\theta}{2}\right)\mathrm{e}^{i\phi}.
$$
La fonction propre normalisée est donc
\begin{equation*}
\chi_+^{(n)} =
\begin{pmatrix}
\cos(\theta/2) \\
\mathrm{e}^{i\phi} \sin(\theta/2)
\end{pmatrix}.
\end{equation*}
